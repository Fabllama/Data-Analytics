```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the file (e.g., CSV format)
data = pd.read_csv('iris/iris.data', delimiter=',', header=None)

# Set the header
data.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class']

# Convert the 'Class' column to numerical values
data['Class_Encoded'] = data['Class'].astype('category').cat.codes

# Create a correlation matrix (including encoded Class column)
corr_matrix = data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class_Encoded']].corr()
```

```python
# Display the correlation matrix
print("Correlation Matrix:")
print(corr_matrix)

# Plot a heatmap for visualization
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
