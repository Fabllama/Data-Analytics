```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y, marker='D', markerfacecolor='red', markeredgecolor='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [3, 7, 1, 8, 5, 10]
plt.bar(categories, values, color='grey')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```
