```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Apple', 'Grape', 'Banana', 'Orange']
sizes = [25, 30, 20]  # Values for each category
remaining = 100 - sum(sizes)  # Calculate the remaining percentage
sizes.append(remaining)  # Add the remaining percentage to the sizes list
colors = ['red', 'purple', 'yellow', 'orange']  # Colors for each slice
explode = (0, 0, 0, 0.1)  # Explode the first slice

# Create the pie chart
plt.figure(figsize=(8, 6))  # Set the figure size
plt.pie(
    sizes,
    explode=explode,    # Highlight
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',  # Show percentages
    shadow=True,        # Add shadow
    startangle=90       # Start angle
)

# Add a title
plt.title('Favorite fruit of 100 students')

# Save the figure
plt.savefig("piechartActivity.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
